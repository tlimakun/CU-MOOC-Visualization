```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
user_info = pd.read_csv('datasets/user_info.csv')
```

```python
user_info.head()
```

```python
user_info['age'] = user_info['yearofbirth'].apply(lambda y: 2020-y)
```

```python
user_info.head()
```

```python
user_info.info()
```

```python
user_info.dropna(inplace=True)
```

```python
user_info.info()
```

```python
user_info.head()
```

```python
# gender count
ax = sns.countplot(user_info['gender'])
sns.despine()

for p in ax.patches:
    ax.annotate(format(p.get_height(), '.1f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points')
```

```python
# education count with gender as legend
ax = sns.countplot(user_info['education'], hue=user_info['gender'])
sns.despine()

for p in ax.patches:
    ax.annotate(format(p.get_height(), '.1f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points')
```

```python
# occupation count with gender as legend
plt.figure(figsize=(15,5))
ax = sns.countplot(user_info['occupation'], hue=user_info['gender'])
sns.despine()

for p in ax.patches:
    ax.annotate(format(p.get_height(), '.1f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points')

plt.xticks(rotation=90)
```

```python
# occupation count with education as legend
plt.figure(figsize=(15,5))
ax = sns.countplot(user_info['occupation'], hue=user_info['education'])
sns.despine()

for p in ax.patches:
    ax.annotate(format(p.get_height(), '.1f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points')

plt.xticks(rotation=90)
plt.legend(loc=1)
```

```python
# residence count with gender as legend
plt.figure(figsize=(15,5))
ax = sns.countplot(user_info['residence'], hue=user_info['gender'])
sns.despine()

for p in ax.patches:
    ax.annotate(format(p.get_height(), '.1f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points')

plt.xticks(rotation=90)
```

```python
# residence count with education as legend
plt.figure(figsize=(15,5))
ax = sns.countplot(user_info['residence'], hue=user_info['education'])
sns.despine()

for p in ax.patches:
    ax.annotate(format(p.get_height(), '.1f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points')

plt.xticks(rotation=90)
```
