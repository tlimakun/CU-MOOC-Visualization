```python
def completed_or_not(row, course_completed_array):
    if row in course_completed_array:
        return 1
    else:
        return 0
```

```python
# add completed column @ course 1
course_1_watch_log_completed = course_1_watch_log.copy()
course_1_watch_log_completed['completed'] = course_1_watch_log_completed['cv_uid'].apply(lambda r: completed_or_not(r, course_1_completed_array))

# add completed column @ course 2
course_2_watch_log_completed = course_2_watch_log.copy()
course_2_watch_log_completed['completed'] = course_2_watch_log_completed['cv_uid'].apply(lambda r: completed_or_not(r, course_2_completed_array))

```
