```
import pandas as pd
import numpy as np
from google.colab import files
```

```
user_log = pd.read_csv('user_activity_log.csv')
```

```
# cv_cid
c1 = 'KTDV6J95MMM8T3NYTKJ8' # Critical Thinking for Business รุ่นที่ 3
c2 = 'OTS26V4DL3U4LE4YHWAF' # ทำอย่างไรให้สินค้า Go Inter ขายได้ขายดี และดัง...
```

```
# user log for each course
c1_user_log = user_log[user_log['cv_cid'] == c1].drop('record_id', axis=1)
c2_user_log = user_log[user_log['cv_cid'] == c2].drop('record_id', axis=1)
```

```
display(c1_user_log.head())
display(c2_user_log.head())
```

```
c1_enrolled_array = c1_user_log[c1_user_log['action_type'] == 'enrolled']['cv_uid'].unique()
c2_enrolled_array = c2_user_log[c2_user_log['action_type'] == 'enrolled']['cv_uid'].unique()
```

```
# check total users enrolled each course
print(len(c1_enrolled_array))
print(len(c2_enrolled_array))
```

```
# get only record of users who have enrolled action_type in each course
c1_user_log = c1_user_log[c1_user_log['cv_uid'].isin(c1_enrolled_array)]
c2_user_log = c2_user_log[c2_user_log['cv_uid'].isin(c2_enrolled_array)]
```

```
# check total users enrolled each course from new dataframe
print(c1_user_log['cv_uid'].nunique())
print(c2_user_log['cv_uid'].nunique())
```

```
# users completed each course array
completed = ['completed_intime', 'completed_late']
c1_completed_array = c1_user_log[c1_user_log['action_type'].isin(completed)]['cv_uid'].unique()
c2_completed_array = c2_user_log[c2_user_log['action_type'].isin(completed)]['cv_uid'].unique()
```

```
# check total users completed each course
print(len(c1_completed_array))
print(len(c2_completed_array))
```

```
def add_completed(user, completed_users):
  if user in completed_users:
    return 1
  else:
    return 0
```

```
c1_user_log['completed'] = c1_user_log['cv_uid'].apply(lambda user: add_completed(user, c1_completed_array))
c2_user_log['completed'] = c2_user_log['cv_uid'].apply(lambda user: add_completed(user, c2_completed_array))
```

```
# check total user completed from completed column of each course
# display head of each dataframe
print(c1_user_log[c1_user_log['completed'] == 1]['cv_uid'].nunique())
print(c2_user_log[c2_user_log['completed'] == 1]['cv_uid'].nunique())

display(c1_user_log.head())
display(c2_user_log.head())
```

```
c1_user_log.info()
c2_user_log.info()
```

```
c1_user_log.to_csv('c1_user_log.csv', index=False)
c2_user_log.to_csv('c2_user_log.csv', index=False)
```
